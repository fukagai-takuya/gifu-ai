```python
import torch
from diffusers import FluxPipeline

pipe = FluxPipeline.from_pretrained(
    "black-forest-labs/FLUX.1-schnell",
    torch_dtype=torch.float16
).to("cuda")
```

```python
prompt = '"Gifu AI Study Group" written using only numbers.'
prompt += 'The letters are constructed from digits (0–9) in various colors, size and rotations.'
prompt += 'Clean white background, vibrant and diverse colors, sharp details, balanced composition.'

image = pipe(
    prompt,
    guidance_scale=0.0,
    num_inference_steps=4,
    max_sequence_length=256,
    generator=torch.Generator("cpu").manual_seed(0)
).images[0]

image.save("Gifu-AI-study-group.png")
```

```python
from PIL import Image
import os

out_dir = "steps"
os.makedirs(out_dir, exist_ok=True)

# GIF 用に画像を溜めるリスト
gif_frames = []

def callback(step: int, timestep: int, latents: torch.FloatTensor):
    with torch.no_grad():
        image = pipe.vae.decode(
            latents / pipe.vae.config.scaling_factor
        ).sample

        image = (image / 2 + 0.5).clamp(0, 1)
        image = image.cpu().permute(0, 2, 3, 1).numpy()

        img = Image.fromarray((image[0] * 255).astype("uint8"))

        # 各 step の PNG 保存（任意）
        img.save(os.path.join(out_dir, f"step_{step:03d}.png"))

        # GIF 用に追加
        gif_frames.append(img)

pipe(
    prompt="a futuristic city at sunset",
    num_inference_steps=8,   # schnell
    callback=callback,
    callback_steps=1
)

# ===== 推論後に GIF を作成 =====
gif_path = "denoising.gif"

gif_frames[0].save(
    gif_path,
    save_all=True,
    append_images=gif_frames[1:],
    duration=200,  # ms / frame
    loop=0
)

print(f"GIF saved to {gif_path}")
```
