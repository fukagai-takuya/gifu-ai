4. FLUX.2 [klein] 4B を使用した Pipeline を用意します

```python
import torch
from diffusers import Flux2KleinPipeline

dtype = torch.bfloat16
pipe = Flux2KleinPipeline.from_pretrained("black-forest-labs/FLUX.2-klein-4B", torch_dtype=dtype)
```
